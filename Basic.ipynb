## People / Time

```python
Chart("dataframe_reduced.csv").mark_area(
    stacked='center',
).encode(
    color=Color('other_person:N',
        scale=Scale(
            range='category20',
        ),
    ),
    x=X('time:T',
        axis=Axis(
            axisWidth=0.0,
            format='%Y',
            labelAngle=0.0,
            tickSize=0.0,
        ),
        scale=Scale(
            nice='month',
        ),
        timeUnit='yearmonth',
    ),
    y=Y('sum(chars):Q',
#         axis=False,
    )
).transform_data(
    calculate=[
#         Formula(field='Direction',
#                        expr='datum.from_me == "True" ? "Sent" : "Recieved"'),
        Formula(field='chars',
               expr='length(datum.content)'),
    ]
    
#     filter='df.known_person'
).configure_cell(
    height=500.0,
    width=800.0,
)
```
